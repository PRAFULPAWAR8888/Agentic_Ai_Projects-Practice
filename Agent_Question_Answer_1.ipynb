```python
from openai import OpenAI
```

```python
openai = OpenAI(
    api_key=userdata.get('GOOGLE_API_KEY'),
    base_url="https://generativelanguage.googleapis.com/v1beta"
)
```

```python
# create a list of messages in the familer Openai format
messages = [{"role":"user","content": "what is 2 +2"}]
```

```python
# And now call it! any problem, head to the Troubleshooting guide
response = openai.chat.completions.create(
    model="gemini-2.0-flash",
    messages=messages,
)

print(response.choices[0].message.content)
```

```python
# and now ask for a question:
question = "Please  propose a hard challanging question to assess someones IQ. Respond only with question."
messages = [{"role":"user", "content":question}]
```

```python
response = openai.chat.completions.create(model="gemini-2.0-flash", messages=messages)

question = response.choices[0].message.content
print(question)
```

```python
messages = [{"role": "user","content":question}]
```

```python
response = openai.chat.completions.create(model="gemini-2.0-flash",messages=messages)
answer = response.choices[0].message.content
print(answer)
```
